```python
def max(list, a=0, b=None, c="x", **kwargs):
    print(f"a: {a}, b: {b}, c: {c}")
    print(f"kwargs: {kwargs}")
    max = list[0]
    max_index = 0
    for i in range(1, len(list)):  # i von 1 - länge-1
        if list[i] > max:
            max = list[i]
            max_index = i
    return max, max_index


m, i = max(list, c="9", d=1)
print(f"m: {m}, i: {i}")
```

```python
def minimax(list):
    min = list[0]
    max = list[0]
    for i in range(1, len(list)):
        if list[i] < min:
            min = list[i]
        elif list[i] > max:
            max = list[i]
    return min, max


minimax(list)
```

```python
def fun(x, w, b=0.0, act=None):
    s = 0
    for i in range(len(x)):
        s = s + x[i] * w[i]
    a = s + b
    if act is not None:
        a = act(a)
    return a


def activate_1(x):
    result = 1 if x > 0 else 0
    return result


def relu(x):
    result = 0 if x < 0 else x
    return result


x1 = [1.4, 0.2]
x3 = [4.7, 1.4]
w = [-0.4, 1.0]
b = 0.4
print(fun(x1, w, b, act=relu))
print(fun(x3, w, b, act=relu))
```

```python
class Decision:
    def __init__(self, w, b=0.0, act=None):
        self.w = w
        self.b = b
        self.act = act if act is not None else lambda x: x

    def predict(self, x):
        return fun(x, self.w, self.b, self.act)


d = Decision([-0.4, 1.0], 0.4, activate_1)
d.predict(x3)
```
