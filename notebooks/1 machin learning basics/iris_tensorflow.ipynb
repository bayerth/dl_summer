```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import tensorflow as tf
import tensorflow.keras as keras
import matplotlib.pyplot as plt

plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(
        iris.data, iris.target, test_size=0.1, random_state=42
        )
```

```python
x = X_train[0]
print(x)
```

```python
xt = tf.constant(x)
print(xt)
```

```python
l = keras.layers.Dense(3, activation="softmax")
```

```python
print(xt.shape)
x_inp = tf.reshape(xt, (1, 4))
print(x_inp.shape)
print(x_inp)
```

```python
y_pred = l(x_inp)
print(y_pred.shape)
y_pred
```

```python
Xt = tf.constant(X_train[:10])
print(Xt.shape)
l(Xt)
```

```python
model = keras.Sequential(
        [
                keras.layers.Dense(
                        8,
                        activation="relu",
                        input_shape=(4,),
                        name="hidden_layer",
                        kernel_initializer="he_normal"
                        ),
                keras.layers.Dense(8, activation="relu", name="hidden_layer2"),
                keras.layers.Dense(3, activation="softmax")],
        name="iris_model"
        )
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model.predict(Xt)
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=15)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_pred_classes = tf.argmax(y_pred, axis=1)
print(y_pred_classes.numpy())
print(y_test)
```

```python
model.evaluate(X_test, y_test)
```

```python
model.save("iris_model.keras")
```

```python
model.layers[1].bias
```

```python
callback = keras.callbacks.EarlyStopping(monitor='loss', patience=4, min_delta=0.0005)
model.fit(X_train, y_train, epochs=50, batch_size=15, callbacks=[callback])
```

```python
inp = [5.1, 3.5, 1.4, 0.2]


def classify(inp):
    x_inp = tf.reshape(tf.constant(inp), (1, 4))
    y_pred = model.predict(x_inp)[0]
    print(y_pred.shape)
    y_class = tf.argmax(y_pred).numpy()
    print(f"Predicted probabilities: {y_pred}")
    print(f"Predicted class: {y_class} ({iris.target_names[y_class]})")
    return y_class


classify(inp)
```
